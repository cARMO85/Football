```python
import pandas as pd
import matplotlib.pyplot as plt
from highlight_text import fig_text
import matplotlib as mpl
from mplsoccer.pitch import Pitch, VerticalPitch
```

```python
#set general use colours
text_color = 'w'
```

```python
#import csv
data = pd.read_csv('/Users/paul/iCloud Drive (Archive)/Downloads/shotmaps.csv')
```

```python
data
```

```python
# create a vertical pitch with half=True
pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc', half=True)

# draw the pitch and set the figure size
fig, ax = pitch.draw(figsize=(10, 10))

# set the facecolor of the figure
fig.set_facecolor('#22312b')

# set up the colors for different outcomes
colors = {'Missed': '#ea6969', 'Saved': '#4e878c', 'Post': '#ad993c'}

# loop through the different outcomes and plot them with the appropriate color
for outcome, color in colors.items():
    df_outcome = data[data['outcome'] == outcome]
    plt.scatter(df_outcome['x'], df_outcome['y'], c=color, s=100, alpha=0.7, label=outcome)

# invert the y-axis to match the orientation of the pitch
plt.gca().invert_yaxis()

# set the title and axis labels
plt.title('Barcelona Shots Vs Juventus', fontsize=24, color=text_color)
plt.xlabel('X', fontsize=16, color=text_color)
plt.ylabel('Y', fontsize=16, color=text_color)

# add the legend
legend = ax.legend(fontsize=16, loc='upper right', facecolor='#22312b', edgecolor='None')
for text in legend.get_texts():
    plt.setp(text, color=text_color)
plt.axis('off')
```
