```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.lines as lines
from mplsoccer.pitch import Pitch
```

```python
# Read in the data
df = pd.read_csv('/Users/paul/iCloud Drive (Archive)/Documents/messibetis.csv')
df
```

```python
df['x'] = df['x']*1.2
df['y'] = df['y']*.8
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*.8
```

```python
fig ,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_type='statsbomb', pitch_color='#3c4d2f', line_color='#c7d5cc')
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

for x in range(len(df['x'])):
    if df['outcome'][x] == 'Successful':
        ax.plot((df['x'][x],df['endX'][x]),(df['y'][x],df['endY'][x]),color='green')
        ax.scatter(df['x'][x],df['y'][x],color='green')
    if df['outcome'][x] == 'Unsuccessful':
        ax.plot((df['x'][x],df['endX'][x]),(df['y'][x],df['endY'][x]),color='red')
        ax.scatter(df['x'][x],df['y'][x],color='red')

fig.set_facecolor('#22312b')

# Create custom lines and labels for the legend
successful_line = lines.Line2D([], [], color='green', marker='o', linestyle='None', label='Successful')
unsuccessful_line = lines.Line2D([], [], color='red', marker='o', linestyle='None', label='Unsuccessful')

# Add the legend to the plot
plt.legend(handles=[successful_line, unsuccessful_line], fontsize=14, facecolor='black', labelcolor='white', edgecolor='white')

# Create title
plt.title('Messi Pass Map vs Real Betis',color='white',size=20)
```
